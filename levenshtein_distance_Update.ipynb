```python
df = df.rename(columns = {'Entity Type' : 'Entity_type', 
                          'Original Watchlist Entity Name' : 'Original_name', 'Test Name Variation' : 'Name_variation'})
```

```python
def column_cleaning(dataframe, column_name, clean_column_name):
    dataframe[f'{clean_column_name}'] = ''
    for i in range(len(dataframe)):
        dataframe[f'{clean_column_name}'][i] = str(dataframe[f'{column_name}'][i]) \
            .lower() \
            .translate(str.maketrans('', '', string.punctuation)) \
            .replace("‘", '') \
            .replace('’', '')
```

```python
column_cleaning(df, 'Original_name', 'Original_name_clean')
column_cleaning(df, 'Name_variation', 'Name_variation_clean')
```

```python
df
```

```python
def remove_numbers(dataframe, column_name, digits_column):
    dataframe[f'{digits_column}'] = ''
    for i in range(len(dataframe)):
        for c in df[f'{column_name}'][i]: 
            # Moving digits to a different column called Name_variation_digits:
            if c.isdigit():
                df[f'{digits_column}'][i] += c
        # Removing digits from the original Name_variation column
        df[f'{column_name}'][i] = df[f'{column_name}'][i].translate(str.maketrans('', '', string.digits))
```

```python
remove_numbers(df, 'Original_name_clean', 'Original_name_digits')
remove_numbers(df, 'Name_variation_clean', 'Name_variation_digits')
```

```python
def column_addition(dataframe, column_name, clean_column_name):
    dataframe[f'{clean_column_name}'] = ''
    for i in range(len(dataframe)):
        dataframe[f'{clean_column_name}'][i] = str(dataframe[f'{column_name}'][i] + ' ')
```

```python
column_addition(df, 'Original_name_clean', 'Original_name_clean1')
column_addition(df, 'Name_variation_clean', 'Name_variation_clean1')
```

```python
def levenshtein_algo(Dataframe, Input1, Input2, Output,Output2,Output3):

    # Creating output column:
    Dataframe[Output] = ''
    Dataframe[Output2] = 0
    Dataframe[Output3] = ''
    aa = 0
    for i in range(len(Dataframe)):
            Dataframe[Output][i] = enchant.utils.levenshtein(Dataframe[Input1][i],Dataframe[Input2][i])
            if Dataframe[Output][i] < 5:
                Dataframe[Output2][i] = 1
                Dataframe[Output3][i] = 'Match'
            else :
                s1 = Dataframe[Input1][i]
                s2 = Dataframe[Input2][i]
                Input1len = len(re.findall(r'\w+', s1))
                Input2len = len(re.findall(r'\w+', s2))
                for j in range (Input1len):
                    k = s1.index(' ');
                    word = s1[0:k];
                    s1 = s1[k:];
                    for l in range (Input2len):
                        k1 = s2.index(' ');
                        word2 = s2[0:k1];
                        s2 = s2[k1:];
                        m = enchant.utils.levenshtein(word,word2)
                        if m == 0:
                            aa = aa + 1
                            s2 = Dataframe[Input2][i]
                if aa > 0:
                    Dataframe[Output2][i] = 1
                    Dataframe[Output3][i] = 'Same Meaning'
                else :
                    Dataframe[Output2][i] = 0
                    Dataframe[Output3][i] = 'Difference is too high'
                aa = 0
```

```python
levenshtein_algo(df, 'Original_name_clean1', 'Name_variation_clean1', 'levenshtein_distance','Score','Result')
```
